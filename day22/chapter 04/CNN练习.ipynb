# 卷积层练习

```python
import torch
import torch.nn as nn

# 随机一个Tensor，形状为 (batch_size=1, in_channels=3, height=32, width=32)
x = torch.randn(1, 3, 32, 32)
print("输入x的shape:", x.shape)

# 创建一个Conv2d层，out_channels是卷积核的个数，in_channels是输入通道数
conv = nn.Conv2d(in_channels=3, out_channels=6, kernel_size=5, stride=1, padding=0)
# 通过卷积层
y = conv(x)
print("卷积后y的shape:", y.shape)
```

# 池化层练习

```python
x = torch.randn(1, 3, 32, 32)
print("输入x的shape:", x.shape)
# 最大值池化
max_pool = nn.MaxPool2d(kernel_size=2, stride=2)
y_max = max_pool(x)
print("最大值池化后y_max的shape:", y_max.shape)

# 平均值池化
avg_pool = nn.AvgPool2d(kernel_size=2, stride=2)
y_avg = avg_pool(x)
print("平均值池化后y_avg的shape:", y_avg.shape)
```
